```python
data_dir = 'drive/MyDrive/your/root/path/Case Presentation 1 Data/'
train_dir = 'Train_Textual/'
test_dir = 'Test_Intuitive/'
valid_dir = 'Validation/'
```

```python
def TXTtoList(dir_path, isValid):
  texts, labels = [], []
  for filename in os.listdir(dir_path):
    text = open(dir_path+filename).read()
    text_list = text.split()
    final_text = ' '.join(text_list)
    texts.append(final_text)
    if isValid:
      ID = filename.split('.')[0].split('_')[1]
      labels.append(int(ID))
    else:
      if filename[0] == 'Y':
        labels.append(1)
      else:
        labels.append(0)
  return texts, labels
```

```python
texts, labels = [], []
dir_path = data_dir + train_dir
for filename in os.listdir(data_dir+train_dir):
  text = open(dir_path+filename).read()
  text_list = text.split()
  final_text = ' '.join(text_list)
  texts.append(final_text)
  if filename[0] == 'Y':
    labels.append(1)
  else:
    labels.append(0)

dir_path = data_dir + test_dir
for filename in os.listdir(dir_path):
  text = open(dir_path+filename).read()
  text_list = text.split()
  final_text = ' '.join(text_list)
  texts.append(final_text)
  if filename[0] == 'Y':
    labels.append(1)
  else:
    labels.append(0)
```

```python
# train_txt, train_label = TXTtoList(data_dir+train_dir, False)
test_txt, test_label = TXTtoList(data_dir+test_dir, False)
valid_txt, valid_label = TXTtoList(data_dir+valid_dir, True)
```

```python
def LISTtoCSV(texts, labels, dir_path, filename):
  data_dict = {"text": texts, "label": labels}

  # 轉為dataframe再透過pandas轉成csv檔
  result_df = pd.DataFrame(data_dict)
  result_df.to_csv(dir_path+'{}.csv'.format(filename), index=False)
  print(result_df)
```

```python
LISTtoCSV(texts, labels, data_dir, 'train')
LISTtoCSV(test_txt, test_label, data_dir, 'test')
LISTtoCSV(valid_txt, valid_label, data_dir, 'valid')
```
